```python
import llb
```

```python
text = "I have 20 flips with 14 heads from a mint coin. What's the bias?"
data = {"num_flips": 20, "num_heads": 14}
targets = ["bias"]
OLLAMA_URL = "http://localhost:11434/api/generate"
OLLAMA_MODEL = "llama3:latest"

posterior = llb.infer(
    text,
    data,
    targets,
    api_url=OLLAMA_URL,
    api_key=None,
    api_model=OLLAMA_MODEL,
    n_models=8,
    mcmc_num_warmup=50,
    mcmc_num_samples=100,
)
```
