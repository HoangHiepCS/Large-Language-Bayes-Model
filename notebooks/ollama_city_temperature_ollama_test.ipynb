# City Temperature Test (Ollama)

This notebook builds a city-temperature prediction test similar to the paper example.
Given day1/day2 training pairs and day1 test temperatures, it predicts day2 test temperatures.

```python
import llb
import numpy as np
```

```python
text = (
    "There is training data for a bunch of different cities on random days, along with the temperature for the following day. \
Then, there is test data for a bunch of different days in those same cities. Predict the temperature on the following days. \
Assume different cities tend to have different weather, and weather tends to be similar on neighboring days."
)

data = {
    "num_cities": 5,
    "num_days_train": 10,
    "num_days_test": 10,
    "day1_temp_train": [
        [61.0, 74.9, 61.5, 73.2, 78.1, 69.8, 55.2, 74.2, 53.9, 75.0],
        [73.0, 78.5, 80.2, 78.9, 82.0, 84.2, 79.0, 81.3, 77.3, 76.0],
        [79.3, 82.1, 82.2, 76.0, 81.2, 84.9, 79.0, 79.6, 81.1, 85.7],
        [81.0, 76.5, 82.0, 78.8, 77.6, 77.1, 76.9, 82.2, 83.7, 78.5],
        [79.8, 78.8, 77.8, 79.6, 76.8, 75.4, 80.4, 75.6, 81.0, 76.2],
    ],
    "day2_temp_train": [
        [61.5, 69.9, 62.0, 73.4, 78.8, 68.8, 53.1, 74.5, 49.4, 76.2],
        [76.4, 78.0, 82.4, 78.9, 80.2, 82.8, 79.4, 79.1, 78.3, 80.0],
        [79.5, 83.8, 77.9, 76.8, 78.5, 85.2, 79.8, 81.6, 80.0, 85.5],
        [80.5, 77.1, 86.2, 78.2, 75.5, 79.1, 78.9, 81.8, 84.2, 82.7],
        [80.8, 78.5, 75.7, 76.1, 74.0, 80.3, 81.6, 77.8, 79.5, 72.8],
    ],
    "day1_temp_test": [
        [69.3, 77.5, 74.8, 65.3, 58.5, 87.0, 54.9, 48.5, 59.4, 72.9],
        [76.5, 48.2, 84.1, 66.6, 85.1, 81.9, 81.0, 77.4, 71.9, 77.7],
        [80.6, 78.9, 79.4, 84.3, 80.5, 81.7, 80.6, 82.7, 82.2, 79.7],
        [82.7, 74.7, 82.1, 77.9, 82.5, 76.2, 81.8, 76.5, 79.1, 82.1],
        [78.4, 75.1, 81.0, 75.0, 74.5, 74.2, 80.5, 76.5, 79.0, 77.8],
    ],
}

targets = ["day2_temp_test"]

posterior = llb.infer(
    text=text,
    data=data,
    targets=targets,
    api_url="http://localhost:11434/api/generate",
    api_key=None,
    api_model="llama3:latest",
    n_models=20,
    mcmc_num_warmup=100,
    mcmc_num_samples=500,
)

pred = np.asarray(posterior["day2_temp_test"], dtype=float)
print("Posterior target keys:", list(posterior.keys()))
print("day2_temp_test sample shape:", pred.shape)
print("Posterior mean (first 10):", np.asarray(np.mean(pred, axis=0)).reshape(-1)[:10])
```
